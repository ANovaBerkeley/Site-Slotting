```python
def contains_pattern(string, pattern):
    return re.search(pattern, string) is not None
```

```python
anonymized_data_cleaned = anonymized_data.copy()
anonymized_data_cleaned = anonymized_data_cleaned.drop("Unnamed: 0", axis=1)
anonymized_data_cleaned = anonymized_data_cleaned.drop("Please read the information above. Are you aware and do you agree with the site policies?", axis=1)
anonymized_data_cleaned = anonymized_data_cleaned.rename(columns={"Name :D": "Name", "Have you confirmed being a site leader with Jacob or Fernanda?": "Site Leader"})
anonymized_data_cleaned["Site Leader"] = [contains_pattern(str(i), r"Yes") for i in anonymized_data_cleaned["Site Leader"]]
anonymized_data_cleaned = anonymized_data_cleaned.iloc[:, :19]
anonymized_data_cleaned = anonymized_data_cleaned[~anonymized_data_cleaned["Site Leader"]]
anonymized_data_cleaned = anonymized_data_cleaned.drop("Site Leader", axis=1)
anonymized_data_cleaned = anonymized_data_cleaned.reset_index(drop=True)
anonymized_data_cleaned.columns = columns
anonymized_data_cleaned[:5]
```

```python
def generate_random_array(n, original_len=58):
    """Generates an array of all false values, then changes `n` random False
    values to be True."""
    false_array = [False] * original_len 
    for _ in range(n):
        index = randint(0, len(false_array) - 1)
        while false_array[index]:
            index = randint(0, len(false_array) - 1)
        false_array[index] = True
    return false_array
```

```python
random_exec = generate_random_array(5)
random_CC = generate_random_array(10)
random_driver = generate_random_array(8)
random_site_leader = generate_random_array(10)
random_spanish = generate_random_array(8)
```

```python
random_site_leader_true_indices = [i for i in range(len(random_site_leader)) if random_site_leader[i]]
random_sites = [["[DCA] Monday 3:25PM - 4:50PM"], ["[DCA] Wednesday 3:25PM - 4:50PM"], ["[Montera] Thursday 3:25PM - 5:40PM"], ["[DeJean] Monday 2:40PM - 4:20PM", "[DeJean] Wednesday 2:40PM - 4:20PM"], ["[King] Tuesday 3:00PM - 4:30PM", "[King] Thursday 3:00PM - 4:30PM"], ["[Rudsdale] Tuesday 1:35PM - 3:30PM (Spanish section)", "[Rudsdale] Thursday 1:35PM - 3:30PM (Spanish section)"], ["[Rudsdale] Wednesday 10:30AM - 1:00PM (Spanish section)"], ["[Longfellow] Monday 4:00PM - 5:30PM", "[Longfellow] Wednesday 2:00PM - 3:30PM"], ["[SquashDrive] Thursday 3:55PM - 5:35PM"], ["[John Henry] Tuesday 3:30PM - 5:30PM", "[John Henry] Thursday 3:30PM - 5:30PM"]]
for i, index in enumerate(random_site_leader_true_indices):
    random_site_leader[index] = random_sites[i]
```

```python
anonymized_data_cleaned_features = anonymized_data_cleaned.copy()
anonymized_data_cleaned_features["Exec"] = random_exec
anonymized_data_cleaned_features["CC"] = random_CC
anonymized_data_cleaned_features["Driver"] = random_driver
anonymized_data_cleaned_features["Site Leader"] = random_site_leader
anonymized_data_cleaned_features["Spanish Speaking"] = random_spanish
anonymized_data_cleaned_features[:5]
```

4. Run randomized matching algorithm on OLD MEMBER data using cleaned DF.
5. After OLD MEMBER pairings are decided, wait until NEW MEMBERS join club.
6. Send out Google Form, which will query NEW MEMBERS:
    - Top 5 sites
    - CC Membership

    - Has a car/able to drive rental

8. Clean CSV file using Python and Pandas:
    - Fix column names
    - Drop unnecessary columns
    - Convert data types (True/False vs 'Yes'/'No', 1/2/3... vs 1st/2nd/3rd, etc.)
9. Run modified randomized matching algorithm on OLD and NEW MEMBER DATA using new cleaned DF.
10. Sites Officer + VPS are the final check for site pairings. It is important to note that these algorithms are tools to help us slot people, rather than doing all the work. This pipeline is a great starting point for site slotting, as creating an initial pairing which is good is pretty difficult.
